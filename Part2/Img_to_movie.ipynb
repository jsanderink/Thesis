```python
# !pip install imageio
import imageio
from os import listdir
from datetime import datetime
from collections import Counter
```

### Scatterplot visuals

```python
listdir('Visuals')
```

```python
filenames_visual = [item for item in listdir('Visuals') if item[:9]=='VISUAL_EM']

# datetime.strptime(string, '%Y-%m-%d %H:%M:%S')

times = sorted([datetime.strptime(item[9:17], '%Y%m%d') for item in filenames_visual])
readable_times = [item.strftime("%d-%m-%Y") for item in times]

Counter(readable_times)
```

```python
from IPython.display import Image
Image(filename='visuals/{}'.format(filenames_visual[-1])) 
```

```python
START = datetime.now()
with imageio.get_writer('plots_4secs_is_a_day.mp4', format='FFMPEG', fps=72) as writer:
    for filename in filenames_visual:
        print('{}/{}'.format(filenames_visual.index(filename),len(filenames_visual)), end="\r")
        image = imageio.imread("Visuals\{}".format(filename))
        writer.append_data(image)
              
print('this took:', datetime.now()-START)
```
